# 聚类

## 从kNN到k-means

当面临一个分类问题，很直觉化的解决方案就是将预测样本与各个已知标签的样本进行比较，即在定义了距离的样本空间中计算距离，将预测样本归为平均距离最近的标签类中。然而这是一种lazy-method，是基于数据的一种解决方法， 没有得到相应的模型。

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=300, centers=4,
                  random_state=0, cluster_std=1.0)
```

```python
plt.figure(figsize=(7,7),dpi=98)
plt.scatter(X[:,0],X[:,1],c=y,s=50,cmap=plt.cm.Spectral);
```

$$
D_k = \frac{1}{n}\sum_i^n \Vert \mathbf{x}-\mathbf{\mu}_i \Vert^2
$$
简单地，把距离定义为二维欧几里得距离。

```python
def Distance(ist_x, dataset):
    data_size = dataset.shape[0]
    diff_matrix = np.tile(ist_x,(data_size,1)) - dataset
    # return euclidean distance
    return (((diff_matrix**2).sum(axis=1))**0.5).sum()/data_size
```

那么对于新的样本数据[1,2]计算其预测分类：

```python
inX = [1,2]
```

```python
print(Distance(inX, X[y==0]))
print(Distance(inX, X[y==1]))
print(Distance(inX, X[y==2]))
print(Distance(inX, X[y==3]))
```

```python
plt.figure(figsize=(7,7),dpi=98)
plt.scatter(X[:,0],X[:,1],c=y,s=30,cmap=plt.cm.Spectral);
plt.scatter(1,2,c=1,s=80,marker='*',cmap=plt.cm.Spectral)
```
